```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler # used to standardize the data to a common range
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
# Import Dataset

dataset = pd.read_csv('/content/diabetes.csv')
dataset.head()
```

```python
dataset.shape
```

```python
dataset.describe()
```

```python
# menghitung jumlah penderita diabetes. 
# 0 'non-diabetic case'
# 1 'diabetic case'
dataset['Outcome'].value_counts()
```

```python
dataset.groupby('Outcome').mean()
```

```python
# memisahkan data dan label(Outcome)

X = dataset.drop(columns = 'Outcome', axis=1)
Y = dataset['Outcome']

print(X,Y)
```

#### Standarisasi Data
Standarisasi data dengan range tertentu bertujuan untuk memudahkan model ML untuk menghasilkan prediksi yang lebih baik.

```python
standard_scaler = StandardScaler()
```

```python
standard_scaler.fit(X)
```

```python
data_standarisasi = standard_scaler.transform(X)
print(data_standarisasi)
```

```python
X = data_standarisasi
Y = dataset['Outcome']
print(X,Y)
```

### Train Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2, stratify=Y)
```

```python
print(X_train.shape, X_test.shape)
```

```python
print(Y_train.shape, Y_test.shape)
```

### Training Model

```python
classifier = svm.SVC(kernel='linear')
```

```python
# training model dengan svm

classifier.fit(X_train, Y_train)
```

### Model Evaluation
menghitung evaluasi score

```python
# menghitung skor akurasi data training (X_train)

X_train_pred = classifier.predict(X_train)
X_train_akurasi = accuracy_score(X_train_pred, Y_train)
print('Skor akurasi data training adalah: ', X_train_akurasi)
```

```python
# menghitung skor akurasi data test (X_test)

X_test_pred = classifier.predict(X_test)
X_test_akurasi = accuracy_score(X_test_pred, Y_test)
print('Skor akurasi data test adalah: ', X_test_akurasi)
```
